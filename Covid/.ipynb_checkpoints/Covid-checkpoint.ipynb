```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from datetime import datetime,timedelta
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression 
#from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv('owid-covid-data.csv')
dataF = data
dataF = dataF.loc[:, ['location', 'date', 'total_cases', 'new_cases']]
dataF = dataF[data['location']=='Ecuador']
print(dataF)
```

```python
plt.figure(figsize=(13,5))
plt.grid(True)
plt.plot(dataF['date'], dataF['total_cases'])
plt.show()
```

```python
dataF['date'] = dataF['date'].map(lambda x: (datetime.strptime(x, '%Y-%m-%d') - datetime.strptime("2020-02-23", '%Y-%m-%d')).days)
```

```python
xtrain, xtest, ytrain, ytest = tts(dataF['date'], dataF['total_cases'], test_size=0.2)
modelo = LinearRegression()
entrenamiento = np.array(xtrain).reshape(-1, 1)
modelo.fit(entrenamiento, ytrain)
```

```python
prediccion = list(range(min(dataF['date']), max(dataF['date'])+50))
pred = modelo.predict(np.array(prediccion).reshape(-1, 1))
plt.figure(figsize=(13,5))
plt.grid(True)
plt.plot(dataF['date'], dataF['total_cases'], '--', label='Datos Reales')
plt.plot(prediccion, pred, '.-', label="Regresión Lineal")
plt.show()
```
