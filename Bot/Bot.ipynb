```python
!pip install selenium
import sys
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import numpy as np
chrome_options = webdriver.ChromeOptions()
#chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('./chromedriver',chrome_options=chrome_options)
```

```python
temas = ['Informatica','Inteligencia Artificial','Software libre','Hacker','Tarjeta de Video']
linkPag1 = []
linkPag2 = []
linkPag3 = []
resultados1 = []

print("### PAGINA 1 ###")
for i in range(len(temas)):
  resultados = []
  driver.get('https://www.significados.com')
  print(driver.title)
  elemento = driver.find_element_by_id('q')
  elemento.send_keys(temas[i])
  elemento.send_keys(Keys.RETURN)

  try:
    main = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "content-left"))
    )
    articles = main.find_elements_by_tag_name("div")
    header = articles[0].find_element_by_tag_name("a")
    print(header.text)
    header.click()
    definicion = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "text-content"))
    )
    definiciones = definicion.find_elements_by_tag_name("p")
    print(definiciones[0].text)
    print(definiciones[1].text)
    print(definiciones[2].text)
    res1 = definiciones[0].text + ". " + definiciones[1].text + ". " + definiciones[2].text
    resultados1 += [res1]
    linkPag1 += [driver.current_url]
  finally:
    #driver.quit()
    print("------")

print("### PAGINA 2 ###")
resultados2 = []
for i in range(len(temas)):
  driver.get('https://www.definicion.de')
  print(driver.title)
  elemento = driver.find_element_by_id('s')
  elemento.send_keys(temas[i])
  elemento.send_keys(Keys.RETURN)

  try:
    main = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "content"))
    )
    articles = main.find_elements_by_tag_name("article")
    header = articles[0].find_element_by_tag_name("a")
    print(header.text)
    header.click()
    definicion = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "definicion-post-box"))
    )
    definiciones = definicion.find_elements_by_tag_name("p")
    print(definiciones[0].text)
    print(definiciones[1].text)
    print(definiciones[2].text)
    res2 = definiciones[0].text + ". " + definiciones[1].text + ". " + definiciones[2].text
    resultados2 += [res2]
    linkPag2 += [driver.current_url]
  finally:
    #driver.quit()
    print("------")
resultados3 = []
print("### PAGINA 3 ###")

for i in range(len(temas)):
  driver.get('https://www.definicionabc.com/')
  print(driver.title)
  elemento = driver.find_element_by_name('s')
  elemento.send_keys(temas[i])
  elemento.send_keys(Keys.RETURN)

  try:
    main = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "caja.caja-textura.caja-620.izquierda.caja-margen-derecho.marron"))
    )
    articles = main.find_elements_by_tag_name("article")
    header = articles[0].find_element_by_tag_name("a")
    print(header.text)
    header.click()
    definicion = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "definicion"))
    )
    definiciones = definicion.find_elements_by_tag_name("p")
    definiciones2 = definicion.find_elements_by_tag_name("h3")
    try:
      print(definiciones[0].text)
      print(definiciones[1].text)
      print(definiciones[2].text)
      print(definiciones[3].text)
      print(definiciones2[0].text)
      print(definiciones2[1].text)
      res3 = definiciones[0].text + ". " + definiciones[3].text
      resultados3 += [res3]
      linkPag3 += [driver.current_url]
    except:
      r3 = definiciones[0].text + ". " + definiciones[1].text + ". " + definiciones[2].text + ". " + definiciones[3].text
      resultados3 += [r3]
      linkPag3 += [driver.current_url]
  finally:
    #driver.quit()
    print("------")
```
